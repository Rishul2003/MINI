```python
import os
import pandas
import numpy
import pickle
import pefile
import sklearn.ensemble as ek
from sklearn.feature_selection import SelectFromModel
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import svm
import sklearn.metrics as metrics
from sklearn.linear_model import LogisticRegression
```

#### Feature Selection

```python
# Feature
X = dataset.drop(['Name','md5','legitimate'],axis=1).values    #Droping this because classification model will not accept object type elements (float and int only)
# Target variable
y = dataset['legitimate'].values
```

#### Data Fitting and choosing the important variables

```python
# Fit an Extra Trees classifier to the data and perform feature selection
extratrees = ek.ExtraTreesClassifier().fit(X, y)
# Use SelectFromModel to select features based on the fitted classifier
model = SelectFromModel(extratrees, prefit=True)
# Transform the original feature matrix to contain only selected features
X_new = model.transform(X)
# Determine the number of selected features
nbfeatures = X_new.shape[1]
```

```python
#Number of important features
nbfeatures
```

```python
#splitting the data (70% - training and 30% - testing)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y ,test_size=0.29, stratify = y)
```

```python
features = []
index = numpy.argsort(extratrees.feature_importances_)[::-1][:nbfeatures]
```

#### All the required features

```python
for f in range(nbfeatures):
    print("%d. feature %s" % (f + 1, dataset.columns[2+index[f]]))
    features.append(dataset.columns[2+f])
```

#### Testing which Classifier will give better result

```python
model = { 
    "DecisionTree": DecisionTreeClassifier(max_depth=10),
    "RandomForest": ek.RandomForestClassifier(n_estimators=50),
    "SVM": svm.SVC(),
    "GradientBoosting": ek.GradientBoostingClassifier(n_estimators=100),
    # "LogisticRegression": LogisticRegression()
}
```

```python
from sklearn.metrics import confusion_matrix

results = {}
for algo in model:
    clf = model[algo]
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    accuracy = clf.score(X_test, y_test)
    print("%s : Accuracy = %s" % (algo, accuracy))
    print("Confusion Matrix for %s:" % algo)
    print(cm)
    results[algo] = {'accuracy': accuracy, 'confusion_matrix': cm}
```

```python
winner = max(results, key=lambda x: results[x]['accuracy'])
print("Using", winner, "for classification, with", len(features), 'features.')
```

#### Saving the machine as classifier.pkl and features to be extracted as features.pkl

```python
# Save the winning model to a file
joblib.dump(model[winner], 'classifier.pkl')

# Save the features list to a file using pickle
with open('features.pkl', 'wb') as f:
    # Serialize the features list and write it to the file
    pickle.dump(features, f)
```
